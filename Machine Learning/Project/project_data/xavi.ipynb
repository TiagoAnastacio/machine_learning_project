```python
from thefuzz import process, fuzz
import numpy as np
import pandas as pd

correct_brands = ['bmw','toyota','audi','ford','skoda','opel','mercedes','hyundai','vw']

a = df[['Brand']].copy()
a['brand_norm'] = a['Brand'].astype(str).str.lower().str.strip()

def top2_matches(x):
    if pd.isna(x) or x == '':
        return [(np.nan, np.nan), (np.nan, np.nan)]
    res = process.extract(x, correct_brands, limit=2, scorer=fuzz.WRatio)
    # guarantee exactly 2 pairs (choice, score)
    res = (res + [(np.nan, np.nan)]*2)[:2]
    return res

top2 = a['brand_norm'].apply(top2_matches)

# unpack into four columns: match1, score1, match2, score2
a[['match1','score1','match2','score2']] = pd.DataFrame(
    [(r[0][0], r[0][1], r[1][0], r[1][1]) for r in top2],
    index=a.index
)

# optional: cleaned brand using the best match above a threshold
THRESH = 90
a['Brand_clean'] = np.where(a['score1'] >= THRESH, a['match1'], a['Brand'])
```

```python
correct_models = {}

for brand in correct_brands:
    models_series = df.loc[df['Brand'] == brand, 'model'].value_counts()
    top_models = models_series[models_series > 50].index.tolist()  # Threshold = 50 aparições
    correct_models[brand] = top_models
    print(f"Brand: {brand}")
    print("Correct models:", top_models)
```
