```python
import random
```

```python
# function of rooling dice
def roll_dice(n):
    results = []
    for i in range(n):
        results.append(random.randint(1,6))
    return results
```

```python
def create_empty_scorecard():
    scorecard = {
        '1' : None,
        '2' : None,
        '3' : None,
        '4' : None,
        '5' : None,
        '6' : None,
        'three_of_kind' : None,
        'four_of_a_kind': None,
        'full_house': None,
        'four_straight' : None,
        'five_straight' : None,
        'yahtzee' : None,
        'chance' : None      
    }
    return scorecard

```

```python
def select_keep(dice):
    print(dice)
    dice_to_keep = input('Which dice to keep?')
    dice_move_forward = []
    dummy_dice = dice.copy() # here we create a copy of the list dice, so we can eliminate items in the future for loop without changing the original list
    dice_to_keep = dice_to_keep.replace(' ', '')
    for i in dice_to_keep:
        i = int(i)
        if i in dummy_dice:      
            dice_move_forward.append(i)
            dummy_dice.remove(i)
        else:
            raise ValueError('Not Valid Selection')
    return dice_move_forward
        
           
      
```

```python
def reroll(dice, kept):
    for i in range(len(dice) - len(kept)):
        kept.append(random.randint(1,6))
    return kept    
```

```python
def has_straight(dice, length):
    dummy_dice = dice.copy()
    dummy_dice = set(dummy_dice)
    dummy_dice = list(dummy_dice)
    dummy_dice.sort()
    is_there_sequence = 0
    for idx, i in enumerate(dummy_dice):
        check_length = list(range(i, i + length, 1))
        if check_length == dummy_dice[idx: idx + length]:
            is_there_sequence += 1
    if is_there_sequence > 0:
        return True
    else:
        return False
```

```python
def evaluate(dice):
    scores = {
        '1' : None,
        '2' : None,
        '3' : None,
        '4' : None,
        '5' : None,
        '6' : None,
        'three_of_kind' : None,
        'four_of_kind': None,
        'full_house': None,
        'four_straight' : None,
        'five_straight' : None,
        'yahtzee' : None,
        'chance' : None      
    }
    total = sum(dice) 
    
    counts = {} #conta quantas vezes x nÃºmero aparece
    for i in dice:
        if i in counts:
            counts[i] += 1
        else:
            counts[i] = 1
            
    for i in range(1, 7):
        scores[str(i)] = dice.count(i) * i #multiplicar o valor do dado por quantas vezes ele aparece

    for value in counts.values():
        if value >= 5:
            scores["yahtzee"] = 50
            scores["four_of_kind"] = total
            scores["three_of_kind"] = total
        if value >= 4:
            scores["four_of_kind"] = total
            scores["three_of_kind"] = total
        elif value >= 3:
            scores["three_of_kind"] = total
    
    full_house = []
    for value in counts.values():
        full_house.append(value)
    if 2 in full_house and 3 in full_house:
        scores["full_house"] = 25
    
    if has_straight(dice,5)==True:
            scores["four_straight"] = 30
            scores["five_straight"] = 40
    elif has_straight(dice,4)==True:
        scores["four_straight"] = 30
    
    scores["chance"] = total
    
    return scores
```

```python
print(evaluate([2,2,2,3,4]))
```
