```python
# %load main.py
from yahtzee import *

def main():
    """Main game loop for Yahtzee."""
    name = input('Enter your name please:')
    print(f"Welcome to Yahtzee, {name}!")
    
    
    # Initialize scorecard
    card = create_empty_scorecard() 

    # Play 13 rounds 
    for i in range(2):
        play_round(card)
        if i != 1:
            print(f'---------------Round: {i+1}-------------- ')
        else:
            print(f'---------------Final Scorecard-------------- ')
        
    
    # Display scorecard after each round
    upper = sum((card[k] or 0) for k in ('1','2','3','4','5','6'))
    bonus = 35 if upper >= 63 else 0
    total = bonus + sum((v or 0) for v in card.values())


    # Calculate final score
    # If upper section total >= 63, bonus = 35
    # --- [Your code here] ---
    
    print(f"\nGame over! Final score: {total}\nWell Done, {name}")

if __name__ == "__main__":
    main()
```
