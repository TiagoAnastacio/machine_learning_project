```python
# Data for testing
grade_a = 15 
grade_b = 21 
grade_c = 8  

# 1. Helper Function: Create 'is_valid' (Returns Boolean)
# YOUR CODE HERE
def is_valid (x):
    if x >= 0 and x <= 20:
        return True
    return False
# 2. Helper Function: Create 'is_passing' (Returns Boolean)
# YOUR CODE HERE
def is_passing(x):
    if x >= 10:
        return True
    return False
# 3. Main Function: Create 'check_scholarship_eligibility' (Composes helpers)
# YOUR CODE HERE
def check_scholarship_eligibility(x):
    if is_passing(x) and is_valid (x):
        return True
    return False
# CALL YOUR FUNCTIONS HERE (e.g., status_a = check_scholarship_eligibility(grade_a))
print(check_scholarship_eligibility(x=grade_a))
print(check_scholarship_eligibility(x=grade_b))
print(check_scholarship_eligibility(x=grade_c))
# Inspect the resulting variables after running this cell
```
